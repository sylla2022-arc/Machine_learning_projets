```python
def plot_weight(alpha = 'None', coefs : list = [], name = 'model_name' ):
    plt.figure(figsize=(6, 3))
    plt.plot(alphas, coefs)
    plt.xscale('log')
    plt.xlabel("Valeurs d'alpha", fontsize = 10)
    plt.ylabel('Poids', fontsize = 10)
    plt.title(f"Régualarisation par {name}", fontsize = 14)
    plt.tight_layout()
    plt.draw()
```

```python
def plot_cross_val_result(alphas = 'None', scores : list = [], model = 'model_name') :
    plt.figure(figsize=(6, 3))
    plt.plot(alphas, scores, linestyle='-')
    plt.xscale('log')
    plt.xlabel("Valeurs d'alpha")
    plt.ylabel('Mean Squared Error', fontsize = 11)
    plt.title(f"Cross-Validated - Mean Squared Error vs Alpha\n pour le modèle {model}", fontsize = 11)
    plt.tight_layout()
    plt.draw()
```

**Data traitement**      
Nous allons d'abord procéder à la detection des outliers de quelques variables :  **total sulfur dioxide** , **chlorides**, **residual sugar** et **sulphates**.

```python
def outlier(df, col = 'col_name'):
    Q1= df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    I = Q3 - Q1
    val_min =  Q1 - 1.5*I
    val_max =  Q3 + 1.5*I
    data_outlier = df[(df[col] < val_min) | (df[col] > val_max)]
    return data_outlier
```

```python
outlier_sulfure = outlier(data, "total sulfur dioxide")
outlier_chloride = outlier(data, "chlorides")
outlier_sugar = outlier(data, "residual sugar")
outlier_sulfate = outlier(data, "sulphates")
```

```python
print(f"Le nombre d'outliers detecté dans la variable total sulfur dioxide est : {outlier_sulfure.shape[0]}")
print(f"Le nombre d'outliers detecté dans la variable chlorides est : {outlier_chloride.shape[0]}")
print(f"Le nombre d'outliers detecté dans la variable residual sugar est : {outlier_sugar.shape[0]}")
print(f"Le nombre d'outliers detecté dans la variable sulphates est : {outlier_sulfate.shape[0]}")
```

**Puis, ramener ces valeurs à la hauteur du troisième quartile**.

```python
outlier_sulfure_index = outlier_sulfure.index.tolist()
outlier_chloride_index = outlier_chloride.index.tolist()
outlier_sugar_index = outlier_sugar.index.tolist()
outlier_sulfate_index = outlier_sulfate.index.tolist()
```

```python
def outlier_treatement(df, index_list : list = [] , col = 'col_name') :
    for ind in index_list :
        df.loc[df.index == ind, col] = df[col].quantile(0.75)
```

```python
outlier_treatement(data, outlier_sulfure_index, col = "total sulfur dioxide")
outlier_treatement(data, outlier_chloride_index, col = "chlorides")
outlier_treatement(data, outlier_sugar_index, col = "residual sugar")
outlier_treatement(data, outlier_sulfate_index, col = "sulphates")
```

**Commentaire**: 

On est assez bon pour prendre la route vers la modélisation.

#### Commentaire :

On va maintenant appliquer les regularisationS LASSO et Ridge pour réduire cette erreur des moyennne carrées.

### <font color="red" id = section_4 > IV. Regularisation par LASSO et optimisation du coef alpha </font>  
La methode LASSO utilise un coefficient pour pondérer les variables, c'est ce coefficient qu'il faut optimiser pour biaiser la prédiction et donc rduire l'erreur. On l'appelle $\alpha$ et sa valeur par défaut est 1.

```python
alphas = np.logspace(-3, 1, 300)
cv_scores = []
kf = KFold(n_splits=5, shuffle=True)

# Boucler sur chaque valeur d'alpha
for alpha in alphas:
    lasso = Lasso(alpha=alpha)
    scores = cross_val_score(lasso, X_train_scaled, y_train, cv=kf, scoring='neg_mean_squared_error')
    cv_scores.append((-1) * np.mean(scores))
```

**Visualisation de l'erreur après regularisation**

```python
plot_cross_val_result(alphas, cv_scores, model = 'Lasso')
```

```python
np.round(min(cv_scores), 4)
```

**Commentaire**

* On a fait plutot un peu mieux car on était à $0.4942$ et on sait retrouver avec $0.4023$.
* On va maintenant observer comment Lasso pondére les caractéristiques.

```python
lasso = Lasso()
coefs = []

for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(X_train_scaled, y_train)
    coefs.append(lasso.coef_)
```

```python
plot_weight(alphas, coefs, name = 'Lasso')
```

#### Commentaire :  
Ce visuel nous explique deux choses importantes :
* Plus la valeur de $\alpha$ devient grande, plus le poids en valeur absolue des caractéristiques dimunue.
* Certaines variables explicatives sont selectionnées très tôt et affectées d'un coefficient bas autrement **Lasso** raméne leur poids vers $0$.

En resumé, ces caratéristiques ne sont pas supprimées mais sont pondérées différemment car **Lasso** les considére comme moins influentes dans la détermination de la qualité du vin.

### <font color="red" id = section_5 > V. Regularisation par Ridge </font>

On va procéder de la même facon. Commençons d'abord par observer comment **Ridge** régule les caractéristiques d'entrées.

```python
alphas = np.logspace(-1, 6, 300) 
```

```python
rigde = Ridge()

coefs = []
for a in alphas:
    rigde.set_params(alpha=a)
    rigde.fit(X_train_scaled, y_train)
    coefs.append(rigde.coef_)
```

```python
plot_weight(alphas, coefs, name = 'Ridge')
```

#### Commentaire :  
On voit bien une différence visuelle entre les deux.  
* Dans le même ordre d'idées, on a les mêmes conclusions avec **Lasso** sur la tendance lorsque $\alpha$ augmente.
* Mais la différence se situe plutôt au niveau du choix autremen dit **Ridge** ne fait aucune sélection des caractéteristiques d'entrées cependant certaines sont plutôt pondérées avec un poids plus bas par rapport à d'autre.

 **Visualisation de MSE après regularisation**

```python
scores = []
for alpha in alphas:
    rigde = Ridge(alpha=alpha)
    cv_scores = cross_val_score(rigde, X_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')
    scores.append((-1) * np.mean(cv_scores))
```

```python
plot_cross_val_result(alphas, scores, model = 'Ridge')
```

```python
np.round(min(scores), 4)
```

**Commentaire**:  
Comme on peut le voir, la régulation a également fonctionné avec **Ridge**.  
En gros, la qualité de la régularisation depend entre autre du type de données avec lequel on travail.
